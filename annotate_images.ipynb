```python
import torch
import os

# Load the pre-trained YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # You can change 'yolov5s' to 'yolov5m', 'yolov5l', or 'yolov5x'

# Define paths for images and output
image_folder = r"D:\swarup_classification\AI_Classification\Scene Classification\test\forest"
output_folder = r"D:\swarup_classification\AI_Classification\Scene Classification\test\forest_annotation"

# Process each image
for img_name in os.listdir(image_folder):
    img_path = os.path.join(image_folder, img_name)
    if img_path.endswith(".jpg") or img_path.endswith(".png"):
        # Run the model on the image
        results = model(img_path)

        # Save results as YOLO annotations (you need to customize this to save in YOLO format)
        results.save()  # saves output images with predictions
        results.xywh[0]  # You can extract the YOLO format annotations from this

        # Save the annotation file (class_id, x_center, y_center, width, height)
        with open(os.path.join(output_folder, img_name.replace(".jpg", ".txt")), "w") as f:
            for result in results.xywh[0]:  # loop through detected objects
                class_id = int(result[5].item())  # Class ID
                x_center = result[0].item()  # x center
                y_center = result[1].item()  # y center
                width = result[2].item()  # width
                height = result[3].item()  # height

                # Write to file in YOLO format
                f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")

```
